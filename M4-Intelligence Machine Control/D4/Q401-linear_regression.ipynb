```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%config InlineBackend.figure_format = "retina"

ar = pd.read_csv("./data/Area_Rental.csv")
print(ar)

print(ar.corr())
print()

ar.plot(x="Area", y="Rental", style="o")
plt.xlabel("mother blood pressure")
plt.ylabel("baby blood pressure")

x = ar[["Area"]]
y = ar["Rental"]

lrm = LinearRegression()
lrm.fit(x, y)

print(lrm.intercept_)
print(lrm.coef_)

predictions = lrm.predict(x)
plt.scatter(x, y, color="black")
plt.plot(x, predictions, color="blue", linewidth=3)
plt.title("Area vs Rental")
plt.xlabel("Area")
plt.ylabel("Rental")

print("\nYou must pay",lrm.predict([[50]]))
#print("\nYou must pay",lrm.predict([[160]]))

print("\nMAE: ", metrics.mean_absolute_error(y, predictions))
print("MSE: ", metrics.mean_squared_error(y, predictions))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y, predictions)))
```
