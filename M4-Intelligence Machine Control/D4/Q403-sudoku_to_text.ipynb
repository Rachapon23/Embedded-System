```python
import cv2
from PIL import Image
import pytesseract
import numpy as np
pytesseract.pytesseract.tesseract_cmd = r'D:\\Tesseract-OCR\\tesseract.exe'

# img = Image.open("./image/test_image.png")
# img.show()

# text_from_img = pytesseract.image_to_string(img)
# print(text_from_img)

# img = cv2.imread("./image/Sodoku_01.jpg")
img = cv2.imread("./image/Sodoku_02.jpg")
# img = cv2.imread("./image/Sodoku_03.jpg")
# img = cv2.imread("./image/Sodoku_04.jpg")
# img = cv2.imread("./image/Sodoku_02z.jpg")
# img = cv2.imread("./image/Sodoku_01.jpg")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_display = img.copy()
edges = cv2.Canny(img_gray, 50, 150)
# cv2.imshow("canny" ,edges)
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, 100, minLineLength=100, maxLineGap=10)
# cv2.imshow("img_gray" ,img_gray)

min_x, min_y, max_x, max_y = img.shape[1], img.shape[0], -1, -1

for line in lines:
    x1, y1, x2, y2 = line[0]
    min_x = min(min_x, x1, x2) 
    min_y = min(min_y, y1, y2)
    max_x = max(max_x, x1, x2) 
    max_y = max(max_y, y1, y2)
    cv2.line(img_display, (x1, y1), (x2, y2), (0,255,0), 2)
    
cv2.line(img_display, (min_x, min_y), (max_x, min_y), (0,0,255), 3)
cv2.line(img_display, (min_x, min_y), (min_x, max_y), (0,0,255), 3)
cv2.line(img_display, (min_x, max_y), (max_x, max_y), (0,0,255), 3)
cv2.line(img_display, (max_x, min_y), (max_x, max_y), (0,0,255), 3)
# cv2.line(img_display, (x1, y1), (x2, y2), (0,255,0), 2)

cv2.imshow("Original", img)
cv2.imshow("Result", img_display)

x_step = (max_x - min_x) / 9
y_step = (max_y - min_y) / 9
offset_x = x_step * 10 / 100
offset_y = y_step * 10 / 100

for row in range(9):
    for col in range(9):
        x_start = int(min_x + (x_step * col) + offset_x)
        x_stop  = int(min_x + (x_step * (col + 1)) - offset_x)
        y_start = int(min_y + (y_step * row) + offset_y)
        y_stop  = int(min_y + (y_step * (row + 1)) - offset_y)
#         print( x_start,x_stop)
        roi_img = img[y_start:y_stop, x_start:x_stop]
#         cv2.imshow("Result-"+str(row)+" "+str(col), roi_img)

        ret, img_bin = cv2.threshold(roi_img, 127, 255, cv2.THRESH_BINARY)
#         cv2.imshow("img_bin" ,img_bin)

        img_rgb = cv2.cvtColor(img_bin, cv2.COLOR_BGR2RGB)
#         cv2.imshow("img_rgb" ,img_rgb)
        
        img_array = Image.fromarray(img_bin)
        text_from_img = pytesseract.image_to_string(img_array, lang="eng", config="--psm 10 --oem 3 -c tessedit_char_whitelist=0123456789")
        
        if(len(text_from_img) == 1):
            print("- ", end="")
        else:
            print(text_from_img[0]+" ", end="")
    print()

cv2.waitKey(0)
cv2.destoryAllWindows()
```
