```python
import cv2
import time
import numpy as np
        
cap = cv2.VideoCapture('./video/save2.avi')
font = cv2.FONT_HERSHEY_TRIPLEX
countGreen = 0
countRed = 0
obj_detector = cv2.createBackgroundSubtractorMOG2()
detect1 = False
detect2 = False
detect3 = False
detect4 = False
isGreen = False
isRed = False
while(True):
    ret, frame = cap.read()

    # Cut the Area, Blur, Chage to Gray  
    belt = frame[50: 475, 80: 400]
    md_blur_belt = cv2.medianBlur(belt,5)
    gray_belt = cv2.cvtColor(belt, cv2.COLOR_BGR2GRAY)
    _, threshold = cv2.threshold(gray_belt,80,255,cv2.THRESH_BINARY)
    
    
    # Mask Circle and Draw 
    circles = cv2.HoughCircles(threshold,cv2.HOUGH_GRADIENT,1,20,
                               param1=50,param2=30,minRadius=30,maxRadius=80)
    
    edge_circle = []
    result = belt.copy()
    
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0,:]:
            if i[2] >= 60 :  # big nut
                cv2.circle(belt,(i[0],i[1]),i[2],(0,255,0),2)
                edge_circle = (i[0],i[1],i[2])
                isGreen = True
                isRed = False
                break
            
            if 40 < i[2] < 60 :
                cv2.circle(belt,(i[0],i[1]),i[2],(0,0,255),2)
                edge_circle = (i[0],i[1],i[2])
                isRed = True
                isGreen = False
        
    if circles is not None:
        if(edge_circle):
            if(abs(int(edge_circle[1]) - int(edge_circle[2])) >= 290):
                detect1 = True
                cv2.circle(result, (i[0],i[1]),   5, (0,0,255), 3)
                cv2.rectangle(
                    belt, (0, 290), 
                    (319, 305),
                    (255, 255, 255), thickness=cv2.FILLED
                )
            if(edge_circle[1] >= 290 and edge_circle[1]  <= 310):
                detect2 = True
                cv2.circle(result, (i[0],i[1]),   5, (0,0,255), 3)
                cv2.rectangle(
                    belt, (0, 290), 
                    (319, 305),
                    (255, 255, 255), thickness=cv2.FILLED
                )
            if(abs(int(edge_circle[1]) - int(edge_circle[2])) >= 350):
                detect3 = True
                cv2.circle(result, (i[0],i[1]),   5, (0,0,255), 3)
                cv2.rectangle(
                    belt, (0, 290), 
                    (319, 305),
                    (255, 255, 255), thickness=cv2.FILLED
                )
            if(edge_circle[1] >= 350 and edge_circle[1]  <= 360):
                detect4 = True
                cv2.circle(result, (i[0],i[1]),   5, (0,0,255), 3)
                cv2.rectangle(
                    belt, (0, 290), 
                    (319, 305),
                    (255, 255, 255), thickness=cv2.FILLED
                )
                
            if((detect1 and detect2) or (detect3 and detect4) ):
                if(isGreen):
                    isGreen = False
                    isRed = False
                    countGreen += 1
                elif(isRed):
                    isGreen = False
                    isRed = False
                    countRed += 1
                detect1 = False
                detect2 = False
                detect3 = False
                detect4 = False

    
    cv2.rectangle(
        belt, (0, 290), 
        (319, 295),
        (0, 0, 255), thickness=cv2.FILLED
    )
    

    cv2.putText(
        belt, str(countGreen),
        (0,60),
        font, 2, (255,255,255)
    )
    
    cv2.putText(
        belt, str(countRed),
        (200,60),
        font, 2, (255,255,255)
    )
    
    # Show Result and Wait Exit
    cv2.imshow('Output-Frame', belt)
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
    
```
